```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data 

```python
europe = pd.read_csv('../Data/Covid19_Europe_20210710.csv', 
                       parse_dates=['Date'],
                       index_col=0, 
                       low_memory=False)
```

```python
vaccination = pd.read_csv("../Data/Vaccination.csv", index_col=0)
```

# Functions

```python
def plot(df, country):
    
    df['NewCases'] = df.ConfirmedCases.diff().fillna(0)
    df.loc[df.CountryName == country, ['NewCases', 'Date']].set_index('Date').plot()
    
    plt.title(country, size=16)
```

```python
def plot_vaccine(df, country):
    
    df.loc[df.location == country, ['date', 'daily_vaccinations']].set_index('date').plot()
```

### Visualization

```python
countries = ['Italy', 'France', 'Germany', 'United Kingdom', 'Denmark', 'Portugal', 'Greece']

for country in countries:
    plot(europe, country)
```

```python
countries = ['Italy', 'France', 'Germany', 'United Kingdom', 'Denmark', 'Portugal', 'Greece']

for country in countries:
    plot_vaccine(vaccination, country)
```
