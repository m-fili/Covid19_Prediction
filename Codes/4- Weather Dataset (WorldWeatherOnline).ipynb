# Merging

```python
files = glob.glob('../Data/Weather_Individual/*.csv')
cols = ['location','date_time', 'HeatIndexC', 'humidity', 'tempC', 'windspeedKmph', 'precipMM', 'DewPointC', 'pressure']
```

```python
weather_all = pd.DataFrame()

for file in files:
    
    # Read individual files
    temp = pd.read_csv(file, parse_dates=['date_time'], usecols=cols)
    
    # Paste individual files
    weather_all = pd.concat([weather_all, temp], axis='rows', sort=False)
    
    # Sort Columns
    weather_all = weather_all[cols]
```

```python
# Save
weather_all.to_csv('../Data/Weather.csv')
```
